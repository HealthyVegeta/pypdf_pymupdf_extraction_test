```python
from langchain.document_loaders import PyMuPDFLoader
from langchain.document_loaders import PyPDFLoader
```

```python
import os
os.environ["OPENAI_API_KEY"] = "sk-"
```

```python
pdf_path = "./paper.pdf"
loader = PyPDFLoader(pdf_path)
documents_pypdf = loader.load_and_split()
print(documents_pypdf[0].page_content)
```

```python
pdf_path = "./paper.pdf"
loader = PyMuPDFLoader(pdf_path)
documents_pymupdf = loader.load()
print(documents_pymupdf[0].page_content)
```
